```python
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
driver = webdriver.Firefox()
```

```python
driver.get('https://www.youtube.com/@VEVO/videos')
```

```python
video_title = driver.find_elements(By.XPATH,'//*[@id="video-title"]')
video_vues = driver.find_elements(By.XPATH,'//*[@id="metadata-line"]/span[1]')
video_dates = driver.find_elements(By.XPATH,'//*[@id="metadata-line"]/span[2]')

videos=[]
for i in range(len(video_dates)):
    video_info={}
    video_info['Title']=video_title[i].text
    video_info['Views']=video_vues[i].text
    video_info['Dates']=video_dates[i].text
    videos.append(video_info)

videos
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(videos,columns=['Title','Views','Dates'])
df
```

```python
df.to_csv('VEVO_videos.csv', index=False, encoding='utf-8')
```

```python
driver.quit()
```

```python
df1=pd.read_csv('VEVO_videos.csv')
df1
```

```python
def viewsToInt(views):
    if views.split()[1]=='k':
        return int(float(views.split()[0].replace(',','.'))*1000)
    elif views.split()[1]=='m':
        return int(float(views.split()[0].replace(',','.'))*1000000)
    else:
        return int(float(views.split()[0].replace(',','.')))
```

```python
df1['Views']=df1['Views'].apply(lambda df:viewsToInt(df))
```

```python
df1
```

```python
print('Most viewed Title on VEVO :')
df1[df1['Views']==df1['Views'].max()]
```

```python
def datesToHour(date):
    if date.split()[-1]=='ans':
        return int(int(date.split()[-2])*365*24)
    elif date.split()[-1]=='mois':
        return int(int(date.split()[-2])*30.4*24)
    else:
        return int(int(date.split()[-2]))
```

```python
df1['Dates_Hours']=df1['Dates'].apply(lambda df1:datesToHour(df1))
```

```python
df1
```

```python
print('Most recent Title :')
df1[df1['Dates_Hours']==df1['Dates_Hours'].min()]
```

```python
print("The YEARS with the most viewed videos on VEVO :")
df1.groupby('Dates').sum()[df1.groupby('Dates').sum()['Views']==(df1.groupby('Dates').sum()['Views']).max()]['Views']
```

```python
def datesToYear(date):
    if date.split()[-1]=='heures':
        return 2023
    elif date.split()[-1]=='mois':
        return 2023
    else:
        return 2023-int(int(date.split()[-2]))
```

```python
df1['Years']=df1['Dates'].apply(lambda df1:datesToYear(df1))
```

```python
df1
```

```python
df2=df1.groupby('Years').sum()['Views']
```
